## Import wordcloud

```python
from wordcloud import WordCloud,STOPWORDS
```

## Open stock file in read only mode

```python
stocks = open('stocks.txt','r',encoding="utf8").read()
```

```python
stocks
```

```python
stopwords=set(STOPWORDS)
```

## Initialize word cloud and apply generate method

```python
wordCloud = WordCloud(background_color='white',max_words=1000,stopwords=stopwords)
wordCloud
wordCloud.generate(stocks)
print('word cloud generated')
```

## import matplotlib and use imshow method to display wordcloud

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(wordCloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## We can also use image mask so that all the words are shown in that particular image mask

```python
import numpy as np
from PIL import Image
alice_mask = np.array(Image.open('alice_mask.png'))
alice_mask
```

```python

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()

```

```python
wordCloud = WordCloud(background_color='black',max_words=1000,mask=alice_mask,stopwords=stopwords)
wordCloud.generate(stocks)
print('generated')
```

## Finally display the word cloud within the image mask. The output looks like this

```python
fig=plt.figure()
fig.set_figwidth(15)
fig.set_figheight(18)
plt.imshow(wordCloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
